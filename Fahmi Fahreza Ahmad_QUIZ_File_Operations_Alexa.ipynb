```python
# Q 2:
"""
Open the 'quiz_files/flower_names.txt' file in this project with Python.
And append the name below to this file:

'Z: Zinnia elegans'

Finally print all the file content.

Hints:
* os
* append mode ('a')
"""

# Q 2:

path = 'flower_names.txt'
new_flower = 'Z: Zinnia elegans'

with open(path, 'a') as file:
    file.write(new_flower + '\n')

with open(path, 'r') as file:
    print(file.read())
# --- your solution here ---

# --------------------------------------------------------------------------------------#
```

```python
# Q 3:
"""
Create a file named 'quiz_files/file_to_delete.txt'.
And add the content below into this file:
'This file will be deleted in the Quiz.'

Hints:
* os
* to create file -> mode='x'
* encoding
"""

# Q 3:

import os

path = 'quiz_files.txt'
content = 'This file will be deleted in the Quiz.'

with open(path, mode='x', encoding='utf-8') as file:
    file.write(content)

print(f"File '{path}' Success.")

# --- your solution here ---


# --------------------------------------------------------------------------------------#
```

```python
# Q 4:
"""
Delete the file 'quiz_files/file_to_delete.txt' you created in Q3.
Add exception handling to handle the case of not finding such a file.

Hints:
* os
* try-except
* to delete -> os.remove()
"""

# Q 4:
import os

path = 'quiz_files.txt'

try:
    os.remove(path)
    print(f"File '{path}' deleted successfully.")
except FileNotFoundError:
    print(f"File '{path}' not found, cannot delete.")

# --- your solution here ---

# --------------------------------------------------------------------------------------#
```

```python
# Q 5:
"""
Get the list of all content (files and folders) in this project directory.
Use this methods:
1- os.listdir()
2- os.scandir()
3- pathlib.Path.iterdir()

"""

# Q 5:

import os
from pathlib import Path

# 1st Way:
# os.listdir()
files_and_folders = os.listdir('.')
print(files_and_folders)
# --- your solution here ---

# 2nd Way
# os.scandir()
with os.scandir('.') as entries:
    for entry in entries:
        print(entry.name)
# --- your solution here ---

# 3rd Way
# Pathlibh
p = Path('.')
for entry in p.iterdir():
    print(entry.name)
# . -> current folder location
# --- your solution here ---

# --------------------------------------------------------------------------------------#
```
